<h3> Life Expectancy (Dependent Variable) - null values</h3>

Since Life expectancy is our target variable and the number of null values are few, rather than assign a value to these observations that might somehow impact the rest of our analysis we can proceed by eliminating such rows containing null values for the life expectancy variable

```python
df = df.dropna(subset=['life_expectancy'])

#verify the precence of null values in 'life_expectancy'
df['life_expectancy'].isna().sum()
```

<h3>Independent varibles - null values</h3>

As for the independent variables, since they are all numeric variables, to fill in the missing data we can replace the null values with the mean value of the column.

```python
independent_variables = list(df.columns)
index_variables = ['country','year','life_expectancy','status']

#removing our index varible for which we don't need to perfom the replacemnt of the null values
[independent_variables.remove(variable) for variable in index_variables if variable in independent_variables]

#fill the null values with the mean of the specific column
for varible in independent_variables:
    df[varible] = df[varible].fillna(df[varible].mean())
```

```python
#checking for null values
df.isnull().sum()
```

Now our dataset has no missing data

<h3>GDP and Population Outliers</h3>

in order to reduce the impact of outliers and make the data distribution more symmetrical, we can realize the logarithm of the following variables.


```python
# Logarithmic transformation
df['gdp'] = np.log1p(df['gdp'])
df['population'] = np.log1p(df['population'])


# Custom function to handle outliers by capping them
def handle_outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return series.clip(lower_bound, upper_bound)

# Apply the function to log-transformed columns
df['gdp'] = handle_outliers(df['gdp'])
df['population'] = handle_outliers(df['population'])
df['life_expectancy'] = handle_outliers(df['life_expectancy'])

custom_palette = sns.color_palette(['#1f77b4', '#ff7f0e'])

# Create subplots for boxplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Boxplot for GDP
sns.boxplot(y=df['gdp'], ax=axes[0], color=custom_palette[0])
axes[0].set_title('GDP')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Log GDP')
axes[0].grid(False)

# Boxplot for Population
sns.boxplot(y=df['population'], ax=axes[1], color=custom_palette[1])
axes[1].set_title('Population')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Log Population')
axes[1].grid(False)

# Boxplot for Life Expectancy
sns.boxplot(y=df['life_expectancy'], ax=axes[2], color='#2ca02c')  # Green color for life expectancy
axes[2].set_title('Life Expectancy')
axes[2].set_xlabel('Value')
axes[2].set_ylabel('Log Life Expectancy')
axes[2].grid(False)

plt.tight_layout()
plt.show()
```

<h2>Cluster Analysis</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
data = df_encoded.copy()

features = data[['life_expectancy', 'adult_mortality', 'infant_deaths', 'alcohol', 
                 'percentage_expenditure', 'hepatitis_b', 'measles', 'bmi', 
                 'polio', 'total_expenditure', 'diphtheria', 'hiv/aids', 'gdp', 
                 'population', 'thinness_5-9_years', 
                 'income_composition_of_resources', 'schooling']]


#standardize
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

#find k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(features_scaled)
    sse.append(kmeans.inertia_)

#elbow graph
plt.figure(figsize=(6, 4))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Numero di Cluster')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()

#cluster
optimal_k = 3  
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
data['Cluster'] = kmeans.fit_predict(features_scaled)

#plot
sns.pairplot(data, hue='Cluster', vars=['life_expectancy', 'gdp', 'schooling'])
plt.show()


cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)
```

Life Expectancy Information discovered

Cluster Distribution:

Cluster 0 (light pink): Most observations in this cluster show relatively high life expectancy, medium GDP, and high schooling level.
Cluster 1 (light purple): This cluster shows a lower life expectancy distribution than cluster 0. GDP is also medium-low, and the level of schooling is more varied, tending toward lower values than cluster 0.
Cluster 2 (dark purple): Observations in this cluster generally have the lowest life expectancy, lowest GDP, and lowest level of schooling among the three clusters.
Relationships among the Variables:

Life Expectancy and GDP: There is a positive correlation between life expectancy and GDP. Nations with higher GDP tend to have higher life expectancy.
Life Expectancy and Schooling: There is a strong positive correlation between life expectancy and schooling level. Nations with a higher level of schooling tend to have higher life expectancy.
GDP and Schooling: There is also a positive correlation between these two variables. Nations with higher GDP tend to have a higher level of schooling.

<H2>Prediction of the Life Expectancy</h2>

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python

X = df.drop('life_expectancy', axis=1)
y = df['life_expectancy']


X_encoded = pd.get_dummies(X, drop_first=True)


X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


model = LinearRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)


r2 = r2_score(y_test, y_pred)
print(f'R^2 Score: {r2}')

# plot regression
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, edgecolors=(0, 0, 0), c='#1f77b4')  
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4, color='#2ca02c')  #linear regression
plt.xlabel('Valori Reali')
plt.ylabel('Valori Predetti')
plt.title('Linear Regression: Life Expectancy')
sns.set_palette('Set2')  
plt.show()
```

thanks to this regression model we are able to predict with 95% accuracy the life expectancy of a country given its characteristics. Such a model could serve as a useful tool for a country to set target values of certain variables and see how these affect life expectancy and plan medium to long term actions accordingly
